```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2

myImg = cv2.imread('example_image.jpg') #read image 
myImg_rgb = cv2.cvtColor(myImg, cv2.COLOR_BGR2RGB)  #turning to grgb

#plotting
plt.figure()
plt.imshow(myImg_rgb)
plt.axis('off')
```

```python
halfImg = cv2.resize(myImg_rgb, None, fx=0.5, fy=0.5)
plt.figure()
plt.imshow(halfImg)
plt.axis('off')
```
