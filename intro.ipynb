```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2

myImg = cv2.imread('example_image.jpg') #read image 
myImg_rgb = cv2.cvtColor(myImg, cv2.COLOR_BGR2RGB)  #turning to grgb

#plotting
plt.figure()
plt.imshow(myImg_rgb)
plt.axis('off')
```

```python
halfImg = cv2.resize(myImg_rgb, None, fx=0.5, fy=0.5)
plt.figure()
plt.imshow(halfImg)
plt.axis('off')
```

```python
r, g, b = myImg_rgb[:,:,0], myImg_rgb[:,:,1], myImg_rgb[:,:,2]

quart = cv2.resize(myImg_rgb, None, fx=0.25, fy=0.25)

gray = cv2.cvtColor(quart, cv2.COLOR_RGB2GRAY) #gray values

plt.figure()
plt.imshow(gray, cmap = 'gray') #this means gray 
plt.axis('off')
```

```python
# Subplot example
plt.figure(7, figsize=(10, 6))
plt.subplot(2, 3, 1)
plt.imshow(myImg)
plt.title('Original Image')
plt.axis('off')
plt.subplot(2, 3, 2)
plt.imshow(quart)
plt.title('Resized Image')
plt.axis('off')
plt.subplot(2, 3, 3)
plt.imshow(r, cmap='gray')
plt.title('Red Channel')
plt.axis('off')
plt.subplot(2, 3, 4)
plt.imshow(gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.subplot(2, 3, 5)
plt.imshow(g, cmap='gray')

plt.title('Green Channel')
plt.axis('off')
plt.subplot(2, 3, 6)
plt.imshow(b, cmap='gray')
plt.title('Blue Channel')
plt.axis('off')
plt.tight_layout()
plt.show()
```
